```python
# Question 3

inconsistent_distance = (df['trip_distance'] <= 0)
dist = inconsistent_distance.value_counts()
print(dist)
inconsistent_fare = (df['fare_amount'] <= 0)
fare = inconsistent_fare.value_counts()
print(fare)

# The inconsistency in these two variables lies in the fact the output returns True for observations which values are not greater than 0

df2 = df.drop(df[(df['trip_distance'] <=0) & (df['fare_amount']<= 0)].index)
df2

# We remove it from the dataset
```

By default a histogram contains 10 bins so the distribution cannot be well appreciated, therefore if we opt to set it to 1000 we are able to appreciate the distribution of trip distance 



```python
# Question 4

df2.hist(column='trip_distance', bins=1000, log=True)
#df2.plot.hist

# By default a histogram contains 10 bins so the distribution cannot be well appreciated, therefore if we opt to set it to 1000 we are able to appreciate the distribution of trip distance 

```

```python
# Question 5

trip_passenger = df2.groupby('passenger_count')['trip_distance'].median()
print(trip_passenger)

# This is the median trip distance per passenger count

trip_passenger.plot(kind='bar')

# We show graphically
```

```python
# Question 6

import numpy as np
df2['payment_type_text'] = np.where(df2['payment_type'] == 2, 'Cash', 'Other')
df2

# We set a new column called payment_type_text, and by using np.where we can easily set 'Cash' when payment_type = 2 and 'Other' for all other possibilities

```

```python
# Question 7

no_tip = df2['tip_amount'] <= 0
cash = df2['payment_type_text'] == 'Cash'
no_tip.corr(cash)

# The correlation between leaving no tip and paying with cash is really high, 0.92
```
